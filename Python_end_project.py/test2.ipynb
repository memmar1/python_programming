```python
def extract():
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"}
    url = "https://www.impactpool.org/search?q=accountant&commit=search"
    r = requests.get(url,headers)
    soup = BeautifulSoup(r.content, "html.parser")
    return soup
c = extract()
```

```python
def transform(soup):
    divs = soup.find_all("div", class_ = "job ip-layout")    
    for item in divs:
        try:
            newpost = item.find("span", class_ = "label label-warning new-job with-margin-right-xs").text
        except:
            newpost = " "
        title = item.find("a").text.strip()
        grade = item.find("div", class_ = "job-grade").text.strip()
        orgnization = item.find("div", class_ = "job-organization").text.strip()
        #close_date = item.find("div", class_= "closing-date soon").text.strip()         """found a challenge here"""
        #duty_station = item.find("i", class_ = "icon fa fa-map-marker").text.strip()    """found a challenge here"""
        #print(newpost, title, grade,orgnization) """for display as am working"""
        Job = { "New_Posts": newpost,
               "Job_Title": title,
               "Level": grade,
               "Company": orgnization
        }
        Joblist.append(Job)
    return
Joblist = []
transform(c)
print(len(Joblist))
```
